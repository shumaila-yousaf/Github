# Number of videos made by each creator

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
file_path = r'D:/pythonpro/Python Videos for Python Pro Channel (Morning) - Pytho Pro.csv'
df = pd.read_csv(file_path)
creator_counts = df['Creator'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=creator_counts.index, y=creator_counts.values, color='lightblue')  # Set color directly
for i, count in enumerate(creator_counts.values):
    plt.text(i, count + 10, str(count), ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.title('Number of Videos Made by Each Creator')
plt.xlabel('Creator')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Monthly uploaded by Mehraj
# Plot graph by seaborn library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
file_path = r'D:/pythonpro/Python Videos for Python Pro Channel (Morning) - Pytho Pro.csv'
df = pd.read_csv(file_path)
df['Date (Published)'] = pd.to_datetime(df['Date (Published)'], dayfirst=True, errors='coerce')
mehraj_data = df[df['Creator'] == 'Mehraj']
mehraj_monthly_count = mehraj_data['Date (Published)'].dt.to_period('M').value_counts().sort_index()
month_order = ['2023-12', '2024-01', '2024-02', '2024-03']
mehraj_monthly_count = mehraj_monthly_count.reindex(month_order, fill_value=0)
sns.color_palette('viridis')  
plt.figure(figsize=(8, 6))
ax = sns.barplot(x=mehraj_monthly_count.index, y=mehraj_monthly_count.values)
plt.title('Number of Videos Made by Mehraj Each Month')
plt.xlabel('Month')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.tight_layout()
for i, count in enumerate(mehraj_monthly_count.values):
    ax.text(i, count + 5, str(count), ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.show()
```

# Weekly number of uploads by mehraj

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
file_path = r'D:/pythonpro/Python Videos for Python Pro Channel (Morning) - Pytho Pro.csv'
df = pd.read_csv(file_path)
df['Date (Published)'] = pd.to_datetime(df['Date (Published)'], dayfirst=True, errors='coerce')
mehraj_data = df[df['Creator'] == 'Mehraj'].copy()  # Use .copy() to avoid SettingWithCopyWarning
mehraj_data['Week'] = mehraj_data['Date (Published)'].dt.isocalendar().week
mehraj_weekly_count = mehraj_data['Week'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=mehraj_weekly_count.index, y=mehraj_weekly_count.values)
plt.title('Weekly Uploads by Mehraj')
plt.xlabel('Week')
plt.ylabel('Number of Uploads')
plt.xticks(rotation=45)
plt.tight_layout()
for i, count in enumerate(mehraj_weekly_count.values):
    ax.text(i, count + 2, str(count), ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.show()
```
